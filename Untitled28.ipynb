```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('heart_.csv')
df.head(10)
```

```python
df.info()
```

```python
print(df.isnull().sum())
```

```python
df['age']=df['age'].astype(str)
df['age']=df['age'].str.replace('years',' ')
df['age']=pd.to_numeric(df['age'], errors='coerce')
print (df['age'].dtype)
```

```python
import numpy as np
df.loc[(df['age']> 100) | (df['age']< 0),'age']=np.nan
```

```python
mean_age=df['age'].mean()
df['age'] = df['age'].fillna(mean_age)
df['age'] =df['age'].astype(int)
```

```python
df.duplicated().sum()
df=df.drop_duplicates()
print(df.duplicated().sum())
```

```python
df['sex'] = df['sex'].astype(str).str.lower().str.strip()
df['sex'] = df['sex'].replace(['male', 'm', '1', '1.0'], 1)
df['sex'] = df['sex'].replace(['female', 'f', '0', '0.0'], 0)
```

```python
print(df.isnull().sum())
```

```python
platelets_mean = df['platelets'].mean()
df['platelets'] = df['platelets'].fillna(platelets_mean)

sodium_mean = df['serum_sodium'].mean()
df['serum_sodium'] = df['serum_sodium'].fillna(sodium_mean)

print(df.isnull().sum())
```

```python
print(df.info())
```

```python
df['smoking'] = df['smoking'].astype(str).str.lower().str.strip()
df['smoking'] = df['smoking'].replace(['yes', 'y', '1', '1.0'], 1)
df['smoking'] = df['smoking'].replace(['no', 'n', '0', '0.0'], 0)
df['smoking'] = pd.to_numeric(df['smoking'], errors='coerce').fillna(0).astype(int)
```

```python
print(df.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation between Features')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='DEATH_EVENT', data=df, palette='Reds')
plt.title('Deaths vs Survivors')
plt.xlabel('Deaths(1) vs Survivors(0)')
plt.ylabel('Cases')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='DEATH_EVENT', kde=True, element='step', palette='magma')
plt.title('Correlation between Age and Death Events')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='sex', hue='DEATH_EVENT', palette='Reds')
plt.title('Correlation between Sex and Death Events')
plt.xlabel('Male(1) vs Female(0)')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))

sns.boxplot(data=df, x='smoking', y='age', hue='DEATH_EVENT', palette='Reds', width=0.6)

sns.stripplot(data=df, x='smoking', y='age', hue='DEATH_EVENT', palette='Reds',
              dodge=True, alpha=0.3, jitter=True, size=4)

plt.title('age' , fontsize=14)
plt.xticks([0, 1], ['non somking', 'smoking'])
plt.xlabel('smoking case')
plt.ylabel('Age')

handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles[0:2], ['Survivors', 'Death'], title='case')

plt.show()
```
